```
from matplotlib import cm
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(x,y):
  return np.sin(x) + 2*np.cos(y)

res = 150

x = np.linspace(-5,5,num=res)
y = np.linspace(-5,5,num=res)

x, y = np.meshgrid(x,y)

z = f(x,y)

fig, ax = plt.subplots(subplot_kw={"projection":"3d"})

surf = ax.plot_surface(x,y,z, cmap=cm.winter)
fig.colorbar(surf)
```

```
fig2, ax2 = plt.subplots()
level_map = np.linspace(np.min(z), np.max(z), num=25)

cp = ax2.contour(x,y,z,levels=level_map,cmap=cm.winter)
fig2.colorbar(cp)
```
