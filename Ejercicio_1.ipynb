```
import matplotlib.pyplot as plt # libreria para graficar
import numpy as np # libreria para manejo de vectores y utilidades matematicas

def f(m,x,b):
  return m*x + b

res = 100
m = -1
b = 5
x = np.linspace(-10.0, 10.0, num=res)
y = f(m,x,b)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')

```

```
def f(x):
  return 2*x**7 - x**4 + 3*x**2 + 4

res = 100

x = np.linspace(-10.0, 10.0, num=res)
y = f(x)

fig, ax = plt.subplots()
plt.xlim(-2, 2)
plt.ylim(-2, 8)
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
def f(x):
  return np.cos(x)

res = 1000

x = np.linspace(-25.0, 25.0, num=res)
y = f(x)

fig, ax = plt.subplots()
plt.xlim(-25, 25)
plt.ylim(-1.2, 1.2)
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
def f(x):
  return 5**x

res = 1000

x = np.linspace(-25.0, 25.0, num=res)
y = f(x)

fig, ax = plt.subplots()
plt.xlim(-5, 5)
plt.ylim(0, 30)
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
def f(x):
  return np.log2(x)

res = 1000

x = np.linspace(0.01, 25.0, num=res)
y = f(x)

fig, ax = plt.subplots()
plt.xlim(-1, 25)
plt.ylim(-5, 5)
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
def H(x):
  y = np.zeros(len(x))
  for idx, x in enumerate(x):
    if x >= 0:
      y[idx] = 1.0
  return y

res = 1000

x = np.linspace(-25.0, 25.0, num=res)
y = H(x)

fig, ax = plt.subplots()
plt.xlim(-25, 25)
plt.ylim(-1.2, 1.2)
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```
